En la siguiente actividad implementaremos dos algoritmos considerados fundamentales para el manejo de matrices:
**Multiplicación de matrices**: Método tradicional para multiplicar dos matrices.
                                         
                                          
                                           
  **Eliminación Gaussiana/Gauss-Jordan**:Algoritmo para resolver sistemas de ecuaciones y encontrar inversas de matrices.
  
  Tambien analizaremos el desempeño de ambos algoritmos considerando el Número de Operaciones realizadas (multiplicaciones y sumas escaleras) y el Tiempo de Ejecución, esto con el objetivo de comprender cómo la eficiencia computacional varía en función del tamaño de la matriz y evaluar el impacto de acceso a la memoria.

**Planteamiento del problema**
El analisis de algoritmos es fundamental para la optimización de recursos computacionalles, en esta práctica buscamos evaluar el desempeño de dos algoritmos de uso común con aplicaicones científicas y de ingeniería, con la multiplicaciones de matrices evaluaremos el número de operaciones y el tiempo de ejecución al incrementar el tamaño de las matrices y con la Eliminación Gaussiana/Gauss-Jordan resolveremos sistemas de ecuaciones lineales para después comparar la eficiencia de ambas.

**1.-Implementación de algortimos**
                                                                         
  Se codificaran las funciones para multiplicar matrices y realizar la Eliminación Gaussiana, asegurando que se cuenten las operaciones (multiplicaciones y sumas escalares).
  

```python
## **Multiplicación de Matrices**
def multiplicacion_matrices(A, B):
    """Multiplicación de matrices tradicional con conteo de operaciones"""
    n = len(A)
    C = [[0 for _ in range(n)] for _ in range(n)]
    multiplicaciones = 0
    sumas = 0
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
                multiplicaciones += 1
                sumas += 1
    return C, multiplicaciones, sumas

## **Eliminación Gaussiana / Gauss-Jordan**
def gauss_jordan(A):
    """Eliminación Gauss-Jordan para resolver sistemas de ecuaciones con conteo de operaciones"""
    n = len(A)
    multiplicaciones = 0
    sumas = 0
    for i in range(n):
        pivot = A[i][i]
        for j in range(n):
            A[i][j] /= pivot
            multiplicaciones += 1

        for k in range(n):
            if k != i:
                factor = A[k][i]
                for j in range(n):
                    A[k][j] -= factor * A[i][j]
                    multiplicaciones += 1
                    sumas += 1
    return A, multiplicaciones, sumas
```

Comparación de resultados:
                                                                         
                                                                          
  Codificaremos las funciones para multiplicar matrices y realizar la Eliminación Gaussiana, asegurando que se cuenten las operaciones (multiplicaciones y sumas escalares).

```python
import time
import numpy as np
import matplotlib.pyplot as plt

sizes = [100, 300, 1000]
tiempos_multiplicacion = []
tiempos_gauss = []
operaciones_multiplicacion = []
operaciones_gauss = []

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    start = time.time()
    _, mult_ops, sum_ops = multiplicacion_matrices(A.tolist(), B.tolist())
    end = time.time()
    tiempos_multiplicacion.append(end - start)
    operaciones_multiplicacion.append((mult_ops, sum_ops))

    start = time.time()
    _, gauss_mult_ops, gauss_sum_ops = gauss_jordan(A.tolist())
    end = time.time()
    tiempos_gauss.append(end - start)
    operaciones_gauss.append((gauss_mult_ops, gauss_sum_ops))
```

Análisis y Discusión de Resultados:

Visualización de datos:
                                                                  
                                                                   
                                                                    
                                                                     
                                                                      
                                                                       
  Generamos gráficas comparativas que muestrarán el tiempo de ejecución de ambos algoritmos en función del tamaño de la matriz.

```python
plt.figure(figsize=(10,5))
plt.plot(sizes, tiempos_multiplicacion, marker='o', label='Multiplicación de Matrices')
plt.plot(sizes, tiempos_gauss, marker='s', label='Eliminación Gaussiana')
plt.xlabel('Tamaño de la matriz (n)')
plt.ylabel('Tiempo de ejecución (s)')
plt.title('Comparación de tiempos de ejecución')
plt.legend()
plt.grid()
plt.show()
```

**Conclusiones:**

Los experimentos que realizamos demostraron que la eficiencia de los algoritmos varían en función del tamaño de la matriz, la multiplicación de matrices presenta un mayor en el tiempo de ejecución a medida que aumenta el tamaño, debido a su complejidad O(n³) mientras que  la Eliminación Gaussiana muestra una tendencia parecida, pero con una variabilidad que depende de la estructura de la matriz de entrada.

Observamos que el acceso a memoria influye de manera considerable en el desempeño de ambos algoritmos, resaltando la importancia de optimizar el acceso a datos contiguos en memoria para mejorar la eficiencia.
